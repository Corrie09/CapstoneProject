```python
import pandas as pd

# Load the dataset (notebook is in EDA/, data is in project_root/data/)
df = pd.read_csv("../data/okcupid_profiles.csv")
```

Looking at the distinct values of certain column 

```python
col = 'essay0'  # replace with your column name
distinct_values = df[col].dropna().unique()
print(f"Distinct values for {col} (n={len(distinct_values)}):", distinct_values)

# counts (including NaN if any)
print(df[col].value_counts(dropna=False))

# number of distinct non-null values
print("nunique:", df[col].nunique(dropna=True))
# ...existing code...
```

```python
# Summary stats
df.describe(include='all').T
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Correlation matrix for numeric vars
numeric_cols = df.select_dtypes(include='number').columns
sns.heatmap(df[numeric_cols].corr(), cmap='coolwarm', center=0)
plt.title("Correlation matrix of numeric variables")
plt.show()
```
